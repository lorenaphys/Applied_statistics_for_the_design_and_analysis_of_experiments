Definimos como categóricos a los factores.

```r
Datos$Especie <- as.factor(Datos$Especie)
Datos$Sustrato <- as.factor(Datos$Sustrato)
Datos$Calcio <- as.factor(Datos$Calcio)
```

```r
Especie <- Datos$Especie
Sustrato <- Datos$Sustrato
Calcio <- Datos$Calcio
```

Definimos el modelo lineal

```r
m1 <- lm(Y ~ Especie * Sustrato * Calcio)
```
